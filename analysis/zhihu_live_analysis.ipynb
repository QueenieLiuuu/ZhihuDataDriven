```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('../spider/ZhihuLiveDB.xlsx')
```

```python
df
```

```python
review_count = df['review_count']
```

```python
print("Statistics of review_count: min={0}  max={1}  median={2}  mode={3}  Q1={4}  Q3={5}".
      format(review_count.min(), review_count.max(), review_count.median(), review_count.mode(), review_count.quantile(0.25), 
             review_count.quantile(0.75)))
```

```python
df = df[df['review_count']>=11]
```

```python
review_score=df['review_score']
```

```python
print('Statistics of review_score:  mean={0}  mode={1}  median={2}  var={3}'.format(review_score.mean(), review_score.mode(), review_score.median(), review_score.var()))
```
